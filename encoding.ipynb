```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Creating dataset
data = {
    "Name": ["Aman","Riya","Rahul","Sneha","Arjun","Pooja","Karan","Neha","Vikram","Anjali","Anshika"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female","Female"],
    "Department": ["CSE","IT","ECE","CSE","ME","IT","CSE","ECE","ME","CSE","CSSE"],
    "City": ["Delhi","Mumbai","Kolkata","Delhi","Chennai","Mumbai","Delhi","Kolkata","Chennai","Delhi","Bhagalpur"],
    "Performance": ["Good","Average","Poor","Good","Average","Good","Good","Average","Poor","Good","Good"]
}

df = pd.DataFrame(data)

print("Dataset Created Successfully:\n")
print(df)
```

```python
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df
```

```python
oh= OneHotEncoder(sparse_output=False)
data = oh.fit_transform(df[['Department']])
name = oh.get_feature_names_out(['Department'])
encoded_df = pd.DataFrame(data ,columns=name)
encoded_df
```

```python
df = pd.concat([df , encoded_df] , axis=1)
df
```

```python
df.drop('Department' , axis=1 , inplace=True)
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good']])
df['Performance Eencoded'] = oe.fit_transform(df[['Performance']])
df
```
