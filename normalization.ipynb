```python
import pandas as pd
import numpy as np

data = {
    'sname':['s1','s2','s3','s4'],
    'height':[162,171,155,165],
    'weight':[63,60,50,62],
    'speed':[5,3,7,2]
}

```

```python
df=pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import Normalizer
```

```python
num=Normalizer()
df_numeric = df[['height','weight','speed']]
data = num.fit_transform(df_numeric)
df=pd.DataFrame(data, columns=df_numeric.columns)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s_num=StandardScaler()
df_numeric = df[['height','weight','speed']]
data = s_num.fit_transform(df_numeric)
df=pd.DataFrame(data, columns=df_numeric.columns)
df
```
